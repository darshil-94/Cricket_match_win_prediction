```python
import pandas as pd
import numpy as np
```

```python
match = pd.read_csv('matches.csv')
```

```python
deliveries = pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
deliveries.head()
```

```python
match.shape
```

```python
deliveries.shape
```

```python
total_score_df = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df.shape
```

```python
total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

```python
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df['team1'].unique()
```

```python
teams = ['Sunrisers Hyderabad',
         'Mumbai Indians', 
         'Royal Challengers Bangalore',
         'Kolkata Knight Riders', 
         'Kings XI Punjab',
         'Chennai Super Kings',
         'Rajasthan Royals',
         'Delhi Capitals']
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df['team2'].unique()
```

```python
match_df.shape
```

```python
match_df = match_df[match_df['dl_applied']==0]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
deliveries_df = match_df.merge(deliveries,on='match_id')
```

```python
deliveries_df=deliveries_df[deliveries_df['inning'] == 2]
```

```python
deliveries_df
```

```python
deliveries_df['Current_score']=deliveries_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
deliveries_df
```

```python
deliveries_df['runs_left'] = deliveries_df['total_runs_x'] - deliveries_df['Current_score']
```

```python
deliveries_df
```

```python
deliveries_df['Balls_left'] = 126 - (deliveries_df['over']*6 + deliveries_df['ball'])
```

```python
deliveries_df
```

```python
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].fillna("0")
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].astype(int)
deliveries_df['player_dismissed'] = deliveries_df['player_dismissed'].values
wickets = deliveries_df.groupby('match_id')['player_dismissed'].cumsum()
deliveries_df['wickets'] = 10 - wickets
deliveries_df
```

```python
deliveries_df.head()
```

```python
deliveries_df['crr'] = (deliveries_df['Current_score']*6)/(120 - (deliveries_df['Balls_left']))
```

```python
deliveries_df
```

```python
deliveries_df['rrr'] = (deliveries_df['runs_left']*6)/deliveries_df['Balls_left']
```

```python
deliveries_df
```

```python
def rrsult(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
deliveries_df['Result'] = deliveries_df.apply(rrsult,axis=1)
```

```python
final_df = deliveries_df[['batting_team','bowling_team','city','runs_left','Balls_left','wickets','total_runs_x','crr','rrr','Result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df['batting_team'].value_counts()
```

```python
final_df
```

```python
final_df['batting_team'] = final_df['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
final_df['batting_team'] = final_df['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
final_df['bowling_team'] = final_df['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
final_df = final_df[final_df['batting_team'].isin(teams)]
final_df = final_df[final_df['bowling_team'].isin(teams)]
```

```python
final_df
```

```python
final_df['batting_team'].unique()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['Balls_left']!=0]
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(drop='first')
```

```python
final_df_new = oe.fit_transform(final_df[['batting_team','bowling_team','city']]).toarray()
```

```python
final_df_2 = np.hstack((final_df[['runs_left','Balls_left','wickets','total_runs_x','crr','rrr']].values,final_df_new))
```

```python
final_df_2 = final_df_2.tolist()
final_df_DataFram = pd.DataFrame(final_df_2)
```

```python
x = final_df_DataFram.iloc[:,:-1]
y = final_df_DataFram.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_teat,y_train,y_test = train_test_split(x,y,test_size=0.3)
```
